```python
import pandas as pd
import matplotlib.pyplot as plt
```

<div class="alert alert-block alert-info">
Conjuto de dados do titanic
</div>

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df
```

<div class="alert alert-block alert-info">
Gráficos dos sobreviventes
</div>

```python
%matplotlib inline
df["Survived"].value_counts(ascending = True).plot.bar(title = "Gráfico de barra vertical de sobreviventes")
plt.xlabel("1 sobreviveu, 0 morreu")
plt.ylabel("quantidade")
```

```python
%matplotlib inline
df["Survived"].value_counts(ascending = True).plot.barh(title = "Gráfico de barra horizontal de sobreviventes")
plt.ylabel("1 sobreviveu, 0 morreu")
plt.xlabel("quantidade")
```

```python
%matplotlib inline
df["Survived"].value_counts().plot.pie(title = "Gráfico de pizza de sobreviventes", autopct = "%.2f")
plt.xlabel("1 sobreviveu, 0 morreu")
plt.ylabel("quantidade")
```
